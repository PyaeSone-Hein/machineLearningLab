```python
df = pd.read_csv('circle_data.csv', delimiter=';') #put sep to create table on csv
#df = pd.read_csv('circle_data.csv', sep=';') #put sep to create table on csv
df.head()
```

```python
x = df['x']
y = df['y']
rad = df.radius
plt.scatter(x,y);
plt.axis('equal');
```

```python
#change color
```

```python
plt.scatter(x, y, c=rad, s=rad*90, cmap = 'seismic') #cmap is color red
plt.axis('equal');
```

```python
number = np.random.randn(10000) #randn is random normal
plt.hist(number, bins = 25); #bins = rands of values (bucket) to group data in histogram
```

```python
#subplot
```

```python
fig, ax = plt.subplots(2,3, figsize = (12, 6))
ax[0, 0].hist(number, bins = 20) #start from 0
ax[0, 1].hist(number, bins = 20) #1st row, 3rd row
ax[1, 0].hist(number, bins = 20) #2nd row, 1st column
```
