# #2. HANDLING MISSING DATA (MISSING VALUES)

```python
# check dataFrame
hr_df
```

```python
# Let’s find the rows that have Null values
# Notice row index 6 has many missing values
hr_df.isnull()
```

```python
# Let’s look at the number of missing values per column
hr_df.isnull().sum()
```

```python
# Drop all rows that contain Null values
# Note that the data size has been reduced by 7 elements from 1470 to 1463
hr_df.dropna()
```

```python
# Note that row 1 is no longer there!
hr_df.dropna().isnull().sum()
```

```python
# Check if we still have missing values
hr_df.isnull().sum()
```

```python
# Explore alternative (smarter) methods to handle missing values
# Reload the raw data using Pandas as follows
hr_df = pd.read_csv('Human_Resources.csv')
hr_df
```

# NEXT LAB #3: ONE-HOT ENCODING

```python
# Calculate the median monthly rate
monthly_rate_median = individual2_df['MonthlyRate'].median()
print(f"The median monthly rate is: {monthly_rate_median}")

# Fill missing values in 'MonthlyRate' with the median
individual2_df.fillna({"MonthlyRate": monthly_rate_median}, inplace=True)

# Confirm that the missing values in 'MonthlyRate' have been filled
display(individual2_df.isnull().sum())
```
